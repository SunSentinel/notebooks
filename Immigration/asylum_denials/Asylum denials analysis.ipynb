```python
import csv
import pandas as pd
import numpy as np
import altair as alt

alt.renderers.enable('notebook')
```

```python
denials_by_month_df = pd.read_csv("miami-asylum-denials.csv", parse_dates=["fymon"])
denials_by_year_df = pd.read_csv("miami-denials-by-year.csv", parse_dates=["fy"])

denials_by_year_df.info()
```

```python
alt.Chart(denials_by_month_df).mark_bar().encode(
    alt.X('fymon:T', axis=alt.Axis(format='%b %Y', title='Month')),
    alt.Y('percent:Q', axis=alt.Axis(title='Percent denied')),
    tooltip=['percent']
).properties(width=600, title="Asylum denials in Miami immigration court").configure_mark(
    color='#761113'
)
```

```python
alt.Chart(denials_by_year_df).mark_bar(size=12).encode(
    alt.X('fy:T', axis=alt.Axis(format='%Y', title='Year')),
    alt.Y('percent:Q', axis=alt.Axis(title='Percent denied')),
    tooltip=['percent']
).properties(width=450, title="Asylum denials in Miami immigration court").configure_mark(
    color='#761113',
)
```
