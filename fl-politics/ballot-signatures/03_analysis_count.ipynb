```python
import csv
import pandas as pd
import numpy as np
import altair as alt
```

```python
initiatives_df = pd.read_csv('Initiatives_Amendments_Revisions - Initiatives_Amendments_Revisions.csv', encoding="ISO-8859-1", parse_dates=["Date"])
initiatives_df = initiatives_df[(initiatives_df["Status"] == "Passed") | (initiatives_df["Status"] == "Defeated")]
initiatives_df.info()
```

```python
def check_sponsor(sponsor):
    state_sponsors = ["Constitution Revision Commission", "Florida Taxation and Budget Reform Commission"]
    if sponsor in state_sponsors:
        return True
    elif sponsor.startswith("The Florida Legislature"):
        return True
    else:
        return False
```

```python
initiatives_df["state_sponsored"] = initiatives_df["Sponsor"].apply(check_sponsor)
initiatives_df.head(30)
```

```python
initiatives_df.to_csv("initiatives_state_designated.csv")
```

```python
initiatives_df.groupby('state_sponsored').size()
```

```python
initiatives_df["election_year"] = initiatives_df["Election Year"].str[:4]
initiatives_df.tail()
initiatives_df.to_csv("initiative_type_per_date.csv")
```

```python
by_year = initiatives_df.groupby(["election_year", "state_sponsored"]).size().to_frame("count").reset_index()
by_year.to_csv("initiative_type_per_year.csv")
```
